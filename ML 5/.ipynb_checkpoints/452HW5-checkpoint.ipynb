```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_classification
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from scipy.stats import skew
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
```

## 1. Before performing any dimensionality reduction, write a program to use k-means clustering on the Madelon dataset. Try the following k values: 4, 8, 16, 32, 64. 

#### 1.1 What preprocessing techniques did you apply, if any?

1. I checked if all variables are numeric, to see whether there are string type data or un-encoded categorical data, because k-means calculates distance between points so numeric data would be more suitable. And dtypes: int64(500) shows that all 500 variables are numeric.
2. I then checked the skewness of variables to see if they are skewed or have severe outliers. It seems like skewness is not an issue in this case.
3. I then checked the scale of features and samples, because K-means clustering is based on the distance between data points and cluster centers, so the scale of the features can affect the clustering results, especially in this case we are not given units. I applied StandardScaler to the dataset to scale/standardize data, since the range of values is not that large, and I want to give more importance to features with larger variances.

```python
madelon = pd.read_csv("madelon.csv")
madelon.head()
```

```python
madelon = madelon.iloc[:,1:]

# Check if variables are numeric
madelon.info()
```

```python
data, labels = fetch_openml('madelon', version=1, return_X_y=True)

# Check the skewness of each variable
skewness = skew(data, axis=0)

# Count the number of skewed variables
num_skewed = sum(abs(skewness) >= 0.5)
print("Number of skewed variables: {}".format(num_skewed))
```

```python
#Check scale of the dataset

# Histogram of mean values for each feature (column)
plt.hist(madelon.mean(0), bins=50)
plt.title('Histogram of Feature Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of mean values for each sample (row)
plt.hist(madelon.mean(1), bins=50)
plt.title('Histogram of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```

Most of the mean values for each feature (column) between 475 and 490, the histogram is apparently right skewed.
The mean values for each sample(row) seems to be normally distributed with 488 as the mean.

```python
scaler = StandardScaler()
madelon = pd.DataFrame(scaler.fit_transform(madelon))

# Define a list of k values
k_values = [4, 8, 16, 32, 64]

# Iterate over the k values and fit the k-means model
sse = []
silhouette = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=5, init = "k-means++")
    kmeans.fit(madelon)
    sse.append(kmeans.inertia_)
    silhouette.append(silhouette_score(madelon, kmeans.labels_))
    # Print the inertia and silhouette score for each k value
    print(f"k={k}, SSE: {kmeans.inertia_}, Silhouette Score: {silhouette_score(madelon, kmeans.labels_)}")

    
madelon.head()
```

#### 1.2 Describe qualitatively: how does the number of clusters affect the performance of the clustering algorithm on the dataset? 

If the number of clusters is too small, then the algorithm might group together distinct clusters, leading to lower accuracy. Moreover, having too few clusters can result in under-segmentation, where data points with different characteristics are grouped together, leading to a loss of information and patterns in the data. It can also lead to oversimplification of the data, making it difficult to extract meaningful insights.

On the other hand, if the number of clusters is too large, then the algorithm might create many small and highly similar clusters that do not provide any additional insights, resulting in over-segmentation. It can also lead to the creation of artificial clusters, where data points that are not actually similar are grouped together.


In this case, as k increases, both inertia and Silhouette Scores decreased. 
As the number of clusters increases, decreasing SSE means the distance between the data points and the centroids of their assigned clusters decreases. This is because the more clusters there are, the more closely the centroids can fit to the data points.
As the number of clusters k increases, the clusters may become more fragmented and less distinct, which can lead to a lower Silhouette Score. And overlapping might happen with smaller clusters, because there is a negative Silhouette Score indicates that the clusters are not well-separated.

#### 1.3 Generate a plot of the number of clusters k (x-axis) versus the sum of squared distance (SSE) between data points and their assigned centroids (y-axis). What appears to be the optimal k from the list of values you used, and why? 

```python
# Plot the SSE vs k
plt.plot(k_values, sse,"-o")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.show()
```

Overall, SSE decreases as k increases. It seems like k=16 is the optimal k because it appears to be the "elbow point" on the plot, where the SSE begins to level off. Although k=32,46 have lower SSE, the slope is not different from where k=16 starts, and more clusters might lead to over-segmentation.

#### 1.4 For k = 8, how did you initialize the set of cluster centroids before running k-means? Rerun k-means again, this time ensuring that the centroids are all different. Does this â€“ and if so, to what extent â€“ affect the final clusters created?


```python
# Iterate over the k values and fit the k-means model
sse_random = []
silhouette_random = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=5, init = "random")
    kmeans.fit(madelon)
    sse_random.append(kmeans.inertia_)
    silhouette_random.append(silhouette_score(madelon, kmeans.labels_))
    # Print the inertia and silhouette score for each k value
    print(f"k={k}, SSE: {kmeans.inertia_}, Silhouette Score: {silhouette_score(madelon, kmeans.labels_)}")
```

```python
# Plot the SSE vs k for random
plt.plot(k_values, sse_random,"-o")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.show()
```

```python
# Compare two initializations
plt.plot(k_values, sse_random,"-o",label="random")
plt.plot(k_values, sse,"-o",label="kmeans++")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.legend()
plt.show()
```

To ensure the centroids are all different, I changed default initialization from k-means++ to random. K-means++ selects cluster centers that are well-separated from each other. And it can speed up convergence. When initialization is random, it choose n_clusters observations (rows) at random from data for the initial centroids. As a result, the choice of initialization method can affect the convergence rate of the k-means algorithm and the quality of the final clustering solution obtained. 

In this case, from the comparison plot, random and kmeans++ do not differ that much, but kmeans++ has lower SSE when k=16. The optimal value of k would not change. From output, random has slightly larger negative values of Silhouette Score. Overall, the impact on the final clusters created was little.

#### 1.5 More generally for any dataset, what evaluation metrics can be used to assess the quality of the clusters produced?

Silhouette Score, which measures how well each data point is assigned to its own cluster, compared to other nearby clusters. Higher values of the Silhouette Score indicate better clustering solutions.

Calinski-Harabasz Index, which measures the ratio of between-cluster variance to within-cluster variance. Higher values of the Calinski-Harabasz Index indicate better clustering solutions.

Davies-Bouldin Index, which measures the similarity between clusters and the dissimilarity between clusters. Lower values of the Davies-Bouldin Index indicate better clustering solutions.

Rand Index, which measures the similarity between the true cluster labels and the predicted cluster labels. Higher values of the Rand Index indicate better clustering solutions.

Adjusted Mutual Information (AMI), which measures the similarity between the true cluster labels and the predicted cluster labels, while adjusting for chance. Higher values of the AMI indicate better clustering solutions.

## 2. The Madelon dataset is high-dimensional, with 500 features per data point. Some of these features might be redundant or noisy, making clustering more difficult. 

#### 2.1 Fit the standardized data with PCA. Then, create a cumulative variance plot â€“ showing the number of components included (x-axis) versus the amount of variance captured (y-axis). Generally, we want to retain at least 75% of the variance. How many components would you decide to keep?

```python
# Fit PCA model
pca = PCA(n_components = 500).fit(madelon)

# Calculate the cumulative sum of explained variances
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Create a plot of cumulative explained variances
plt.plot(cumulative_variance)
plt.hlines(0.75, 0, 500, linestyles='dashed', colors='r')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.title('Cumulative Variance Plot for Madelon Dataset')
plt.show()
```

The Cumulative explained variance is concave shaped, suggesting that the rate of increase in explained variance is slowing down as more principal components are added, which is expected.

```python
# How many components to keep
n_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) > 0.75) + 1
print(f"Number of principal components to retain: {n_components}")
```

#### 2.2 Perform PCA with your selected principal components.

#### 2.2.1 Plot the transformed data on a graph with the first two principal components as the axes i.e. x = PC 1, y = PC 2.

```python
# select the columns with the first two principal components as the axes
X_pca = PCA(n_components = 277).fit(madelon).transform(madelon)
pca_filtered = X_pca[:, :2]

# plot the first two principal components on a graph 
plt.scatter(pca_filtered[:, 0], pca_filtered[:, 1])
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('Transformed Data (First Two Principal Components)')
plt.show()
```

#### 2.2.2 Plot the original data on a graph with the two original variables that have the highest absolute combined loading for PC 1 and PC 2 i.e. maximizing |loading PC1| + |loading PC2|.

```python
# Get the combined values for PC 1 and PC 2 and Compute their absolute combined loading values
loading_pc1 = PCA(n_components = 277).fit(madelon).components_[0]
loading_pc2 = PCA(n_components = 277).fit(madelon).components_[1]

abs_pc1 = np.abs(loading_pc1)
abs_pc2 = np.abs(loading_pc2)
sum_abs = abs_pc1 + abs_pc2

# Sort the variables by the sum of absolute combined loading values
sorted_vars = np.argsort(sum_abs)[::-1]

# Select the two variables with the highest sum of absolute combined loading values
var1, var2 = sorted_vars[:2]

# Extract these two variables from the original data
data_pc1 = madelon[var1]
data_pc2 = madelon[var2]

# Plot the original data on a scatter plot
plt.scatter(data_pc1, data_pc2)
plt.xlabel(f'Variable {var1}')
plt.ylabel(f'Variable {var2}')
plt.title('Original Data with PC1 and PC2 Variables')
plt.show()
```

## 3. Now, we will run k-means clustering on the transformed data from the previous problem.

### 3.1 Why is it a good idea to do this, especially for high-dimensional datasets? Name a number of reasons.

It can reduce dimensionality. we can reduce the computational cost of k-means clustering and avoid the curse of dimensionality. High-dimensional data can make it difficult to find meaningful clusters because the distances between points become very large, and the volume of the space increases exponentially, making it difficult to separate clusters.

Reducing the dimensionality of the data also leads to better cluster separation, which improves the separation of clusters by reducing the correlation between features and highlighting the most important features that are responsible for the most variance in the data. This can help k-means clustering to identify more meaningful and well-separated clusters.

It also reduces noise, as it helps to remove noise from the data by reducing the influence of low-variance dimensions. This can result in better clustering results, especially in high-dimensional datasets where the noise can be a significant problem.

### 3.3 Again, create a scatter plot of PC 1 (x-axis) versus PC 2 (y-axis) for all of the transformed data points. Label the cluster centers and color-code by cluster assignment for the first 5 iterations of k = 32. Can you see the algorithm begin to converge to optimal assignments?

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

# Cluster assignment for the first 5 iterations of k=32
for i in range(5):
    kmeans = KMeans(n_clusters=32, random_state=5, init="k-means++", max_iter=i+1)
    kmeans.fit(X_pca)
    labels = kmeans.labels_
    centers = kmeans.cluster_centers_
    
    axs[i].scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
    axs[i].set_title(f"Iteration {i+1}")
    axs[i].set_xlabel("PC 1")
    axs[i].set_ylabel("PC 2")

plt.show()
```

Yes, algorithms begin to converge to optimal assignments as colors(labels) are getting distinct, which is a good indication. As the number of iterations increases, the K-means algorithm updates the cluster centroids based on the current assignments of data points to clusters.
