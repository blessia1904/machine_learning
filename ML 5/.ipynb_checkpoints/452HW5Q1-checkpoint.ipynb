## 1. Before performing any dimensionality reduction, write a program to use k-means clustering on the Madelon dataset. Try the following k values: 4, 8, 16, 32, 64.

```python
madelon = pd.read_csv("madelon.csv")
madelon.head()
```

```python
madelon = madelon.iloc[:,1:]

# Check if variables are numeric
madelon.info()
```

```python
data, labels = fetch_openml('madelon', version=1, return_X_y=True)

# Check the skewness of each variable
skewness = skew(data, axis=0)

# Count the number of skewed variables
num_skewed = sum(abs(skewness) >= 0.5)
print("Number of skewed variables: {}".format(num_skewed))
```

```python
#Check scale of the dataset

# Histogram of mean values for each feature (column)
plt.hist(madelon.mean(0), bins=50)
plt.title('Histogram of Feature Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()

# Histogram of mean values for each sample (row)
plt.hist(madelon.mean(1), bins=50)
plt.title('Histogram of Sample Means')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```

Most of the mean values for each feature (column) between 475 and 490, the histogram is apparently right skewed.
The mean values for each sample(row) seems to be normally distributed with 488 as the mean.

```python
scaler = StandardScaler()
madelon = pd.DataFrame(scaler.fit_transform(madelon))

# Define a list of k values
k_values = [4, 8, 16, 32, 64]

# Iterate over the k values and fit the k-means model
sse = []
silhouette = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=5, init = "k-means++")
    kmeans.fit(madelon)
    sse.append(kmeans.inertia_)
    silhouette.append(silhouette_score(madelon, kmeans.labels_))
    # Print the inertia and silhouette score for each k value
    print(f"k={k}, SSE: {kmeans.inertia_}, Silhouette Score: {silhouette_score(madelon, kmeans.labels_)}")

    
madelon.head()
```

#### 1.2 Describe qualitatively: how does the number of clusters affect the performance of the clustering algorithm on the dataset? 

If the number of clusters is too small, then the algorithm might group together distinct clusters, leading to lower accuracy. On the other hand, if the number of clusters is too large, then the algorithm might create many small and highly similar clusters that do not provide any additional insights.
In this case, as the number of clusters increases, decreasing SSE means the distance between the data points and the centroids of their assigned clusters decreases.
As the number of clusters k increases, the clusters may become more fragmented and less distinct, which can lead to a lower Silhouette Score.

#### 1.3 Generate a plot of the number of clusters k (x-axis) versus the sum of squared distance (SSE) between data points and their assigned centroids (y-axis). What appears to be the optimal k from the list of values you used, and why? 

```python
# Plot the SSE vs k
plt.plot(k_values, sse,"-o")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.show()
```

Overall, SSE decreases as k increases. It seems like k=16 is the optimal k because it appears to be the "elbow point" on the plot, where the SSE begins to level off. Although k=32,46 have lower SSE, the slope is not different from where k=16 starts, and more clusters might lead to over-segmentation.

#### 1.4 For k = 8, how did you initialize the set of cluster centroids before running k-means? Rerun k-means again, this time ensuring that the centroids are all different. Does this â€“ and if so, to what extent â€“ affect the final clusters created?


```python
# Iterate over the k values and fit the k-means model
sse_random = []
silhouette_random = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=5, init = "random")
    kmeans.fit(madelon)
    sse_random.append(kmeans.inertia_)
    silhouette_random.append(silhouette_score(madelon, kmeans.labels_))
    # Print the inertia and silhouette score for each k value
    print(f"k={k}, SSE: {kmeans.inertia_}, Silhouette Score: {silhouette_score(madelon, kmeans.labels_)}")
```

```python
# Plot the SSE vs k for random
plt.plot(k_values, sse_random,"-o")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.show()
```

```python
# Compare two initializations
plt.plot(k_values, sse_random,"-o",label="random")
plt.plot(k_values, sse,"-o",label="kmeans++")
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Distance (SSE)')
plt.legend()
plt.show()
```

To ensure the centroids are all different, I changed default initialization from k-means++ to random. K-means++ selects cluster centers that are well-separated from each other. And it can speed up convergence. When initialization is random, it choose n_clusters observations (rows) at random from data for the initial centroids. As a result, the choice of initialization method can affect the convergence rate of the k-means algorithm and the quality of the final clustering solution obtained. 

In this case, from the comparison plot, random and kmeans++ do not differ that much, but kmeans++ has lower SSE when k=16. The optimal value of k would not change. From output, random has slightly larger negative values of Silhouette Score. Overall, the impact on the final clusters created was little.

#### 1.5 More generally for any dataset, what evaluation metrics can be used to assess the quality of the clusters produced?

Silhouette Score, which measures how well each data point is assigned to its own cluster, compared to other nearby clusters. Higher values of the Silhouette Score indicate better clustering solutions.

Calinski-Harabasz Index, which measures the ratio of between-cluster variance to within-cluster variance. Higher values of the Calinski-Harabasz Index indicate better clustering solutions.
